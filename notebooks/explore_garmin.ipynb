# Garmin — Data Exploration

Uses the `garminconnect` package to fetch data from Garmin Connect.

Requires `GARMIN_USERNAME` and `GARMIN_PASSWORD` in `.env`.

```python
import os
from dotenv import load_dotenv
from garminconnect import Garmin

load_dotenv("../.env")
client = Garmin(os.environ["GARMIN_USERNAME"], os.environ["GARMIN_PASSWORD"])
client.login()
print("Logged in")
print(client.)
```

## Activities

```python
# Fetch recent activities
activities = client.get_activities(0, 20)  # (start, limit)
print(f"{len(activities)} activities")
```

```python
import polars as pl

df_activities = (
    pl.DataFrame(activities)
    .select(["activityId", "activityName", "activityType", "startTimeLocal", "duration", "distance", "calories"])
)
df_activities
```

## Steps

```python
from datetime import date, timedelta

today = date.today()
steps = client.get_steps_data(today.isoformat())
print(steps)
```

```python
# Steps over the last 30 days
rows = []
for i in range(30):
    d = (today - timedelta(days=i)).isoformat()
    daily = client.get_daily_steps(d, d)
    if daily:
        rows.extend(daily)

pl.DataFrame(rows)
```

## Sleep

```python
sleep = client.get_sleep_data(today.isoformat())
sleep
```

## Heart Rate

```python
hr = client.get_heart_rates(today.isoformat())
hr
```

## Body Battery

```python
battery = client.get_body_battery(today.isoformat())
battery
```

```python
# Daily wellness summary
import json


rows = []
for i in range(1):
    d = (today - timedelta(days=i)).isoformat()
    s = client.get_stats(d)
    print(json.dumps(
        s,
        sort_keys=True,
        indent=4,
        separators=(',', ': ')
    ))
    if s:
        rows.append({
            "date":               s.get("calendarDate"),
            "steps":              s.get("totalSteps"),
            "distance_m":         s.get("totalDistanceMeters"),
            "calories":           s.get("totalKilocalories"),
            "active_calories":    s.get("activeKilocalories"),
            "resting_hr":         s.get("restingHeartRate"),
            "avg_stress":         s.get("averageStressLevel"),
            "sleep_seconds":      s.get("sleepingSeconds"),
            "floors_ascended":    s.get("floorsAscended"),
        })

df_wellness = pl.DataFrame(rows)
df_wellness
```
